```python

import numpy as np
import matplotlib.pyplot as plt

# Задаем исходные данные
x = np.array([1, 2, 3, 4, 5]) # независимая переменная
y = np.array([2, 3, 4, 5, 6]) # зависимая переменная

# Вычисляем коэффициенты линейной аппроксимации (a и b)
a, b = np.polyfit(x, y, deg=1)

# Создаем функцию для линейной аппроксимации
def linear_approximation(x):
    return a * x + b

# Создаем график и отображаем исходные данные и линейную аппроксимацию
plt.scatter(x, y) # отображение точек данных
plt.plot(x, linear_approximation(x), color='red') # отображение линейной аппроксимации
plt.xlabel('x')
plt.ylabel('y')
plt.title('Линейная аппроксимация')
plt.grid(True)
plt.show()
```
